```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
# map tophat / WGS, use pcawg
wgs_pcawg = pd.read_csv("WGS_PCAWG.96.csv")
tophat_star = pd.read_csv("tophat_star_fpkm_uq.v2_aliquot_gl.tsv", sep="\t")
pcawg_specimen_histology = pd.read_excel("pcawg_specimen_histology_August2016_v9.xlsx")
```

```python
features = tophat_star['feature'][:500]
tophat = []

for feature in tqdm(features):
    vals = tophat_star[tophat_star['feature'] == feature]
    uuids = vals.columns.tolist()[1:]
    for uuid in uuids:
        row = {"feature": feature, "uuid": uuid, "value": vals[uuid].values[0]}
        tophat.append(row)

tophat_df = pd.DataFrame(tophat)
tophat_df.head()
```

```python
tophat_df.columns = ["UserId", "ItemId", "Rating"]
tophat_df.head()
```
